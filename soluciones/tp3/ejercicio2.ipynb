# Implementación de red utilizando el Método Configuracional

## (a) 

Para generar un muestreo aleatorio de la distribucion de grado, vamos a seguir una idea que se detalle en este paper: https://arxiv.org/pdf/0706.1062.pdf (visto en https://stats.stackexchange.com/questions/173242/random-sample-from-power-law-distribution).

Básicamente la idea consiste en convertir el generador pseudoaleatorio en un generador que obecece una distribución tipo ley de potencia.

```python
import random
random.seed(42)
```

```python
# Hiperparámetros 
x_min = 5
```

```python
def random_sample_powerlaw(gamma):
    r = random.random()
    x_smp = x_min * (1 - r) ** (-1 / (gamma - 1))
    return x_smp
```

```python
# Construimos las listas de grados de los nodos
N = [10**3,10**4,10**5]
gamma = [2.5,3,3.5]
k_samples = {}
for gamma_ in gamma:
    sample = {}
    for n_ in N:
        sample_n = []
        for i in range(n_):
            sample_n.append(round(random_sample_powerlaw(gamma_)))
        sample[n_] = sample_n
    k_samples[gamma_] = sample
```

## (d) Construcción de las redes

Primero hay que chequear que el número total de grados es par. Si no lo es, tenemos que modificar la distribución y restar o sumar un enlace a algún nodo del grafo. 

```python
k_dist = k_samples[2.5][1000]
print(sum(k_dist) % 2 == 0)
if sum(k_dist) %2 != 0:
    k_dist[-1] +=1
print(sum(k_dist) % 2 == 0)    
```

Lo que vamos a hacer es una implementación similar a la utilizada en Networkx (https://networkx.org/documentation/networkx-1.10/_modules/networkx/generators/degree_seq.html#configuration_model), en la que construyen una stublist (una lista en la que se repite el $i$ nodo $k_i$ veces):

    # build stublist, a list of available degree-repeated stubs
    # e.g. for deg_sequence=[3,2,1,1,1]
    # initially, stublist=[1,1,1,2,2,3,4,5]
    # i.e., node 1 has degree=3 and is repeated 3 times, etc.
Notemos que esta construcción permite la existencia de autoenlaces y enlaces múltiples. Esto después se puede corregir según sea el caso, eliminando estos enlaces específicos.

Para construir la lista de enlaces, sólo es necesario hacer un _shuffle_ de la stublist, e ir agregando tuplas (i,j) dentro de una lista de enlaces, eliminando luego estos dos elementos de la stublist.

```python
def build_graph(k_dist):
    k_dist_ = k_dist
    if sum(k_dist) %2 != 0:
        k_dist_[-1] +=1
    stublist = []
    for i in range(len(k_dist_)):
        for j in range(k_dist_[i]):
            stublist.append(i)
    edgelist = []
    random.shuffle(stublist)
    while stublist:
        i = stublist.pop()
        j = stublist.pop()
        edgelist.append((i,j))
    return edgelist
```

```python
gamma_25_n_1000 = build_graph(k_samples[2.5][1000])
gamma_25_n_10000 = build_graph(k_samples[2.5][10000])
gamma_25_n_100000 = build_graph(k_samples[2.5][100000])
```

```python
gamma_30_n_1000 = build_graph(k_samples[3][1000])
gamma_30_n_10000 = build_graph(k_samples[3][10000])
gamma_30_n_100000 = build_graph(k_samples[3][100000])
```

## (e) Gráfico de $k_{nn}(k)$ 

Tenemos que construir las listas de vecinos de cada nodo. Para ello vamos a usar una estructura diccionario.

Para hacer el cálculo, necesitamos saber cuántos y cuáles nodos del grafo tienen grado $k$. 

```python
from collections import defaultdict
```

```python
def build_neighbor_list(graph):
    neighbor_list = defaultdict(set)
    for i,j in graph:
        neighbor_list[i].add(j)
        neighbor_list[j].add(i)
    return neighbor_list
```

```python
def get_k_list(k_dist): # Esta función devuelve una lista que tiene en el sitio i una lista de los nodos que tienen grado i
    k_list = []
    for i in range(max(k_dist)):
        k_i = []
        for ii in range(len(k_dist)):
            if k_dist[ii]==i:
                k_i.append(ii)
        k_list.append(k_i) 
    return k_list
```

```python
def knn(k,k_dist,k_list,neighbor_list):
    if k==0:
        return 0
    knn = 0
    nodes_k = k_list[k]
    for node in nodes_k:
        knn_i = 0
        for neighbor in neighbor_list[node]:
            knn_i += k_dist[neighbor]
        knn_i /= k
        knn += knn_i
        knn /= len(nodes_k)
    return knn
```

```python
nl_gamma25_n1000 = build_neighbor_list(gamma_25_n_1000)
klist_gamma25_n1000 = get_k_list(k_samples[2.5][1000])
print('Ready 1000')

nl_gamma25_n10000 = build_neighbor_list(gamma_25_n_10000)
klist_gamma25_n10000 = get_k_list(k_samples[2.5][10000])
print('Ready 10000')

nl_gamma25_n100000 = build_neighbor_list(gamma_25_n_100000)
klist_gamma25_n100000 = get_k_list(k_samples[2.5][100000])
print('Ready 100000')
```

```python
knn_gamma25_n1000 = []
for k in range(max(k_samples[2.5][1000])):
    knn_gamma25_n1000.append(knn(k,k_samples[2.5][1000],klist_gamma25_n1000,nl_gamma25_n1000))
    
knn_gamma25_n10000 = []
for k in range(max(k_samples[2.5][10000])):
    knn_gamma25_n10000.append(knn(k,k_samples[2.5][10000],klist_gamma25_n10000,nl_gamma25_n10000))
    
knn_gamma25_n100000 = []
for k in range(max(k_samples[2.5][100000])):
    knn_gamma25_n100000.append(knn(k,k_samples[2.5][100000],klist_gamma25_n100000,nl_gamma25_n100000))

# Plot
    
fig,ax = plt.subplots(1,3,figsize=(21,7))
plt.suptitle('$\\gamma = 2.5$',size=20)

ax[0].plot(knn_gamma25_n1000,'o')
ax[0].set_title('$N=10^3$',size=18)

ax[1].plot(knn_gamma25_n10000,'o')
ax[1].set_title('$N=10^4$',size=18)

ax[2].plot(knn_gamma25_n100000,'o')
ax[2].set_title('$N=10^5$',size=18)

for ax_ in ax:
    ax_.set_xscale('log')
    ax_.set_yscale('log')
    ax_.set_xlabel('$k$',size=18)
    ax_.set_ylabel('$k_{nn}(k)$',size=18)
    ax_.grid()
    
plt.tight_layout()
```
