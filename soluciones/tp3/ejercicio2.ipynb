## (a) 

Para generar un muestreo aleatorio de la distribucion de grado, vamos a seguir una idea que se detalle en este paper: https://arxiv.org/pdf/0706.1062.pdf (visto en https://stats.stackexchange.com/questions/173242/random-sample-from-power-law-distribution).

Básicamente la idea consiste en convertir el generador pseudoaleatorio en un generador que obecece una distribución tipo ley de potencia.

```python
import random
random.seed(42)
```

```python
# Hiperparámetros 
x_min = 5
```

```python
def random_sample_powerlaw(gamma):
    r = random.random()
    x_smp = x_min * (1 - r) ** (-1 / (gamma - 1))
    return x_smp
```

```python
# Construimos las listas de grados de los nodos
N = [10**3,10**4,10**5]
gamma = [2.5,3,3.5]
k_samples = {}
for gamma_ in gamma:
    sample = {}
    for n_ in N:
        sample_n = []
        for i in range(n_):
            sample_n.append(round(random_sample_powerlaw(gamma_)))
        sample[n_] = sample_n
    k_samples[gamma_] = sample
```

## (b) Verificación de las muestras generadas

### Bineado lineal

```python
from matplotlib import pyplot as plt
import matplotlib
matplotlib.rcParams['axes.labelsize'] = 18
matplotlib.rcParams['xtick.labelsize'] = 14
matplotlib.rcParams['ytick.labelsize'] = 14
```

```python
fig,ax1 =  plt.subplots(1,3,figsize=(21,7))

plt.suptitle('$\\gamma = 2.5$',size=20)

ax1[0].hist(k_samples[2.5][1000],bins=10,density=True)
ax1[0].set_title('$N = 10^3$',size=18)

ax1[1].hist(k_samples[2.5][10000],bins=100,density=True)
ax1[1].set_title('$N = 10^4$',size=18)

ax1[2].hist(k_samples[2.5][100000],density=True)
ax1[2].set_title('$N = 10^5$',size=18)

for ax in ax1:
    ax.set_xlabel('$k$')
    ax.set_ylabel('$P(k)$')
    ax.grid()
plt.tight_layout()
```
